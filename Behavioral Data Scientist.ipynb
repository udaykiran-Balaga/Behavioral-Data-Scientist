# Performance scorecard WEEK 1

```python
import matplotlib.pyplot as plt

# Data for individuals and categories
individuals = ['Neha', 'Amit', 'Sachin']
categories = ['Contribution', 'Initiative', 'Bottleneck Analysis']
scores = [
    [8, 9, 9],
    [8, 8, 7],
    [8, 8, 7]
]

# Create a horizontal bar chart with increased figure size
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2
index = range(len(categories))

# Plot individual scores
for i, ind in enumerate(individuals):
    ax.barh([x + i * bar_width for x in index], scores[i], bar_width, label=ind)
    for j, score in enumerate(scores[i]):
        ax.text(score + 0.1, j + i * bar_width, str(score), ha='center', va='center')

# Add labels and title
ax.set_xlabel('Scores')
ax.set_ylabel('Categories')
ax.set_title('Performance Scorecard - Week 1')
ax.set_xlim(0,12)
ax.set_yticks(index)
ax.set_yticklabels(categories)
ax.legend()

# Invert the y-axis to show the highest score at the top
plt.gca().invert_yaxis()

# Show plot
plt.show()
```
